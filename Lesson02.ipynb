```
#Task1
import numpy as np

matrix = np.random.randint(low=1, high=51, size=(4, 4))
sum_rows = np.sum(matrix, axis=1);
sum_cols = np.sum(matrix, axis=0);
total_mean = np.mean(matrix)
print("Исходный массив 4x4: ")
print(matrix)

print(f"Суммы по строкам: {sum_rows}")
print(f"Суммы по столбцам: {sum_cols}")
print(f"Общее среднее значение: {total_mean:.2f}")
```

```
#Task2
import numpy as np

matrix = np.arange(15).reshape(3, 5)
print("Исходный массив:\n", matrix)

matrix_transpose = np.transpose(matrix)
print("\nТранспонированный массив:\n", matrix_transpose)

reshaped_matrix = matrix.reshape(5, 3)
print("\nМассив после reshape:\n", reshaped_matrix)
```
