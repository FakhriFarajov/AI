```python
#Task3
import pandas as pd
data = pd.read_csv('products_data.csv')
table = pd.DataFrame(data)
table
```

```python
#Task4
import pandas as pd
data = pd.read_csv('work_hours_data.csv')
table = pd.DataFrame(data)
table
```

```python
total_by_employee = table.groupby('Имя')['Часы'].mean()
total_by_employee
```

```python
total_by_project = table.groupby('Проект')['Часы'].mean()
total_by_project
```

```python
total_hours = table.groupby('Имя')['Часы'].sum()
the_most_hardworking_guy = total_hours.idxmax()
the_most_hardworking_guy
```
